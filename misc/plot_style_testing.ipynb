```python
cmap = mpl.colormaps['inferno']
x = np.linspace(0,12,100)

mydata = [np.sin(x+0/(2*np.pi)), 0.6*np.sin(x+0/(2*np.pi)), -0.4*np.sin(x+0/(2*np.pi))+0.3]

num_lines = len(mydata)
for i,y in enumerate(mydata):
    plt.plot(x,y,color=cmap(i/num_lines))
```
